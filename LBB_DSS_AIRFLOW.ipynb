### Membuat Fungsi Data Cleansing

```python
def clean_df(df):
    df['issue_d'] = pd.to_datetime(df['issue_d'], format='%d/%m/%Y', errors='coerce')

    cat_column = ['home_ownership', 'income_category', 'term', 'application_type', 'interest_payments', 'loan_condition', 'grade', 'region', 'purpose']
    df[cat_column] = df[cat_column].astype('category')

    return df
```

### Filtering

```python
# 1. Mengambil waktu paling terupdate dari database
last_update = pd.read_sql_query("""
                  SELECT issue_d 
                  FROM transactions
                  ORDER BY issue_d DESC
                  LIMIT 1
                  """, con = conn)

last_timestamp = last_update['issue_d'].to_string(index=False)
last_timestamp
```

```python
def fetch_clean():
    files = glob.glob("data_input_loan_2014/*.csv")

    for file in files:
        # mengambil transaksi terkini dari database
        last_update = pd.read_sql_query("""
                    SELECT issue_d 
                    FROM transactions
                    ORDER BY issue_d DESC
                    LIMIT 1
                    """, con = conn)
        last_timestamp = last_update['issue_d'].to_string(index=False)

        # membaca dataframe
        trx = pd.read_csv(file)
        # membersihkan dataframe
        trx_clean = clean_df(trx)

        # filtering dataframe
        trx_clean_update = trx_clean[trx_clean['issue_d'] > last_timestamp]
        
        # jika dataframe terisi, maka masukkan ke dalam df_list
        if trx_clean_update.shape[0] != 0:
            df_list.append(trx_clean_update)
    
    return df_list
```

```python
def df_to_db(df_list):
    database = 'db/transactions.db'
    conn = sqlite3.connect(database)

    # iterasi dataframe
    for df in df_list:
        df.to_sql(name = 'transactions',
                con = conn,
                if_exists = 'append',
                index = False)
        
        periode = df['issue_d'].dt.to_period('M').unique()[0]
        print(f'Berhasil Menambahkan: {periode}')
```

```python
# code here
dfs_read = fetch_clean()

dfs_read[0].head()
```
